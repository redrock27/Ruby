```python
import numpy as np
from scipy.optimize import linprog
```

This assignment has three parts.

 **Part-1**
 Optimization of
 min  z= x1 + x2 + x3 + x4

	    x1 + 2x2 -x3 - x4  =1
   -	x1 - 5 x2 + 2x3 + 3x4 =1

     xi ≥ 0    for all   i     
     
      Where number of variables=n=4 and number of constrains=m=2


```python
# Coefficients for the objective function
c = [1, 1, 1, 1]  # Coefficients for x1, x2, x3, x4

# Coefficients for the inequality constraints
A_eq = [
    [1, 2, -1, -1],   # Coefficients for the first equation
    [-1, -5, 2, 3]    # Coefficients for the second equation
]
b_eq = [1, 1]  # Right-hand side of the equations

# Bounds for each variable (xi ≥ 0)
x_bounds = (0, None)

# Define bounds for all variables
bounds = [x_bounds] * 4  # Since we have 4 x variables

# Solving the linear programming problem
result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')

# Display the results
if result.success:
    print("RESULT:")
    print("Optimal value:", result.fun)
    print("Optimal solution (x1, x2, x3, x4):", result.x)
else:
    print("Optimization failed:", result.message)
```

**Part-2**

Here number ofconstrains is same(m=2) but number variables is increasing by increment of 10 ie.(4,10,20,30,40,50)

```python
# Function to solve the linear programming problem
def solve_lp(n):
    # Objective function coefficients (minimize the sum of variables)
    c = np.ones(n)  # Coefficients for x1, x2, ..., xn

    # Create constraint matrices
    A_eq = np.zeros((2, n))

    # Fill the first constraint: x1 + 2x2 - x3 - x4 = 1
    A_eq[0, 0] = 1
    A_eq[0, 1] = 2
    A_eq[0, 2] = -1
    A_eq[0, 3] = -1

    # Fill the second constraint: -x1 - 5x2 + 2x3 + 3x4 = 1
    A_eq[1, 0] = -1
    A_eq[1, 1] = -5
    A_eq[1, 2] = 2
    A_eq[1, 3] = 3

    # The rest of the variables (from x5 to xn) will not affect the equations.
    # Set them to 0 in the constraints
    A_eq[0, 4:n] = 0
    A_eq[1, 4:n] = 0

    # Right-hand side of the equations
    b_eq = [1, 1]

    # Variable bounds (xi ≥ 0 for all i)
    bounds = [(0, None)] * n  # Non-negativity constraint for all variables

    # Solving the linear programming problem
    result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')

    # Display the results
    if result.success:
        print(f"n={n}: Optimal value: {result.fun}, Optimal solution: {result.x}")
    else:
        print(f"n={n}: Optimization failed: {result.message}")

# List of n values to test
n_values = [4, 10, 20, 30, 40, 50]

# Run the optimization for each value of n
for n in n_values:
    solve_lp(n)

print("CONCLUSION:")
print("Here changing number of variable does not change optimal value")
```

**Part-3**
In this part variables are changing by increment of 10 i.e n=(4,10,20,30,40,50)
and constrains are changing by increment of 4 i.e. m=(2,6,10,14,) and also take processing time.

```python
import time

# Function to solve the linear programming problem
def solve_lp(n, m):
    # Objective function coefficients
    c = np.ones(n)  # Coefficients for x1, x2, ..., xn

    # Create constraint matrices
    A_eq = np.zeros((m, n))
    # Fill the first two constraints manually
    A_eq[0, 0] = 1
    A_eq[0, 1] = 2
    A_eq[0, 2] = -1
    A_eq[0, 3] = -1

    A_eq[1, 0] = -1
    A_eq[1, 1] = -5
    A_eq[1, 2] = 2
    A_eq[1, 3] = 3

    # Adding additional constraints randomly for m > 2
    for i in range(2, m):
        A_eq[i, :] = np.random.randint(-5, 5, n)  # Random coefficients
        A_eq[i, np.random.randint(0, n)] = 0  # Ensure at least one coefficient is zero

    # Right-hand side of the equations
    b_eq = [1] * m  # For simplicity, we set all RHS to 1

    # Variable bounds (xi ≥ 0 for all i)
    bounds = [(0, None)] * n  # Non-negativity constraint for all variables

    # Timing the LP solving process
    start_time = time.time()
    result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')
    end_time = time.time()
      # Compute elapsed time
    elapsed_time = end_time - start_time

    # Display the results
    if result.success:
        print(f"n={n}, m={m}:")
        print(f"  Problem Instance: Minimize z = {' + '.join(['x' + str(i+1) for i in range(n)])} under constraints:")
        for i in range(m):
            print(f"    Constraint {i+1}: " + " + ".join(f"{A_eq[i, j]}*x{j+1}" for j in range(n)) + f" = {b_eq[i]}")
        print(f"  Optimal value: {result.fun}")
        print(f"  Optimal solution: {result.x}")
        print(f"  Time taken: {elapsed_time:.6f} seconds\n")
    else:
      print(f"n={n}, m={m}: Optimization failed: {result.message}")

# List of n values and m values to test
n_values = [4, 10, 20, 30, 40, 50]
m_values = [2, 6, 10, 14]

# Run the optimization for each combination of n and m
for n in n_values:
    for m in m_values:
        solve_lp(n, m)
```
